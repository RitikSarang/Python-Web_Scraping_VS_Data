```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.vijaysales.com/mobiles-and-tablets"
r = requests.get(URL)
r
```

```python
def getHTML(url,path):
    with open(path,'w', encoding='utf-8') as fs:
        fs.write(r.text)
```

```python
getHTML(URL,"testHTML.html")
```

```python
with open("testHTML.html",'r') as f:
    html_doc = f.read()

soup = BeautifulSoup(html_doc, 'html.parser')

data = {
    "title" : [],
    "offer_price":[],
    "offer_dicount":[],
    "offer_avail":[],
    "no_cost_emi":[]
}
max_length = max(len(data["title"]), len(data["offer_price"]), len(data["offer_dicount"]))

#Title Extracted
title = soup.find_all("h2",attrs={"class":"BcktPrdNm_"})
for items in title:
    data["title"].append(items.text.strip())

#Offer_price Extracted
offer_price = soup.find_all("span",attrs={"class" : "Prdvsprc_"})
for items in offer_price:
    data["offer_price"].append(items.text.strip())

#Offer_dicount Extracted
offer_discount = soup.find_all("span",attrs={"class":"ofrnm_"})
for items in offer_discount:
    data["offer_dicount"].append(items.text.strip())

#Offer_available_bool Extracted
offer_avail = soup.find_all("span",attrs={"class":"ofravlble"})
for items in offer_avail:
    data["offer_avail"].append(items.text.strip())

#no_cost_emi Extracted
no_cost_emi = soup.find_all("div",attrs={"class":"BcktPrdemi h54"})
for items in no_cost_emi:
    data["no_cost_emi"].append(items.text.strip())
    
#links Extracted
navigatable_links = soup.find_all("a")
"""for items in navigatable_links:
    data["navigatable_links"].append(items.text.strip())"""

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data.items()]))
df
```

```python
# Example: Change text in `text_column` where `column_with_nan` is NaN
df.loc[df['offer_dicount'].isna(), 'offer_avail'] = 'Offer Unavailable'

df.loc[df["offer_dicount"].isna(),'offer_dicount']="0% OFF"
df
```
